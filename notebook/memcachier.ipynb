```python
appid_map = {}
seen = {}

for pm_idx in range(1, 7):
    res = []
    dst = f'/data/zhenyus/tconvert_data/memcachier-trace-long-2day/pair/pm{pm_idx}_merged.csv'
    with open(dst) as f:
        for i, line in enumerate(f):
            r = line.split(',')
            t, appid, cmd, key, size = r[0], r[1], r[2], r[3], r[4]
            if appid not in appid_map:
                appid_map[appid] = len(appid_map)
            res.append([int(t), appid_map[appid], cmd, key, int(size)])
            if i%1000000 == 0:
                print(i)
#             if i == 100000:
#                 break
    # reverse of the requests
    res1 = []
    
    size_map = {}
    for i, r in enumerate(res[::-1]):
        appid = r[1]
        cmd = r[2]
        key = r[3]
        size = r[4]
        if cmd == 'Set' and size != 0:
            #update size map
            size_map[(appid, key)] = size
        else: # cmd == 'Get':
            if size == 0:
                if (appid, key) in size_map:
                    size = size_map[(appid, key)]
                else:
                    continue
            res1.append((r[0], key, size, appid))
        if i%1000000 == 0:
            print(i)

    with open(f'/data/zhenyus/tconvert_data/memcachier-trace-long-2day/pair/pm{pm_idx}.tr', 'w') as f_out:
        for i, r in enumerate(res1[::-1]):
            key = (r[1], r[2])
            if key not in seen:
                seen[key] = len(seen)
            f_out.write(f'{r[0]} {seen[key]} {r[2]} {r[3]}\n')
            if i%1000000 == 0:
                print(i)
```

```python
# spc1-like
seen = {}
page_size = 4096

dst = f'/data/zhenyus/tconvert_data/arc/spc1likeread.lis'
with open(dst) as f:
    with open(f'/data/zhenyus/tconvert_data/arc/spc1likeread.tr', 'w') as f_out:
        for i, line in enumerate(f):
            r = line.split(' ')
            key, size = int(r[0]), int(r[1])
            size *= page_size
            key = (key, size)
            if key not in seen:
                seen[key] = len(seen)
            f_out.write(f'{i} {seen[key]} {size}\n')
            if i%1000000 == 0:
                print(i)
```
