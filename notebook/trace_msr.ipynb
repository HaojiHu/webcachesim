```python
from glob import glob
import subprocess
import tqdm
```

```python
# spc1-like
seen = {}
page_size = 4096

dst = f'/data/zhenyus/tconvert_data/arc/spc1likeread.lis'
with open(dst) as f:
    with open(f'/data/zhenyus/tconvert_data/arc/spc1likeread.tr', 'w') as f_out:
        for i, line in enumerate(f):
            r = line.split(' ')
            key, size = int(r[0]), int(r[1])
            size *= page_size
            key = (key, size)
            if key not in seen:
                seen[key] = len(seen)
            f_out.write(f'{i} {seen[key]} {size}\n')
            if i%1000000 == 0:
                print(i)
```

```python
seen = {}
in_file = '/data/zhenyus/webcachesim/trace/msr_big_200m.tr'
out_file = '/data/zhenyus/webcachesim/trace/msr_rehash_200m.tr'

with open(in_file) as f:  
    with open(out_file, 'w') as f_out:
        for i, line in enumerate(f):
            r = line.split(' ')
            key, size, f1, f2 = int(r[1]), int(r[2]), int(r[3]), int(r[4])
            key = (key, size)
            if key not in seen:
                seen[key] = len(seen)
            f_out.write(f'0 {seen[key]} {size} {f1} {f2}\n')
            if i%1000000 == 0:
                print(i)
```
