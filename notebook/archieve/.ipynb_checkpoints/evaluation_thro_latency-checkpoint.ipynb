```python
%matplotlib inline
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from glob import glob
import pathlib
import common

FONT_SIZE = 15
plt.rc('font', size=FONT_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=FONT_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=FONT_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=FONT_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=FONT_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=FONT_SIZE)    # legend fontsize
plt.rc('figure', titlesize=FONT_SIZE)  # fontsize of the figure title
matplotlib.rc('lines', linewidth=3)
```

```python
df = pd.DataFrame({'no_cache': [10, 20, 30], 'Apache Traffic Server': [4,5,6], 'GDBT': [1,2,3], 'traces': ['Akamai HK', 'Akamai US', 'Wikipedia']})
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
df.plot.bar(x='traces', y=['no_cache', 'Apache Traffic Server', 'GDBT'], ax=ax)
plt.title('FAKE bandwidth reduction')
plt.ylabel('bandwidth consumption GB/hour')
plt.show()
```
