```python
trace = 'wc400m.tr'

df = common.load_reports()
filter_idx = (
    (df.trace_file == trace) 
    & (df.uni_size == 0)
    & (~df.cache_type.isin(['Inf', 'BeladySample'])) 
    & (25000000000 <= df.cache_size)
    & (df.cache_size <= 400000000000)
    & (df.n_warmup == 80000000)
    & (df.cache_type == 'GDBT')
    & (~pd.isna(df.miss_decouple))
    & (df.objective == 'byte_hit_rate')
    & (df.version == 'double_penalty')
    & (df.num_threads == '1')
    & (df.sample_rate == '256')
    & (df.batch_size == '100000')
    & ((df.cache_type != 'LRUK') | (df.k == '4'))
)

df1 = df[filter_idx].copy()
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
forget_windows = df1.forget_window.drop_duplicates()
for f in forget_windows:
    grp = df1[df1.forget_window == f]
    grp = grp.sort_values(by=['cache_size'])
    ax.semilogx(grp.cache_size/1e9,
              grp[objective], 
              label=f'{float(f)/1e6} million', 
              marker='*', 
              markersize=12,
              basex=2)
plt.legend()
plt.xlabel('cache size / GB')
plt.ylabel('byte hit rate')
plt.title(f'{trace_map[trace]} byte hit rate for different forget window')
plt.show()
```

```python
color_map = {
    'Belady': 'k',
#     'BeladySample': '#000080',
    'GDBT': 'b',
    'LR': 'r',
    'LRUK': 'g',
    'LFUDA': 'c',
    'GDSF': 'm',
    'LRU': 'y',
    'S4LRU': '#99ccff',
    'Hyperbolic': 'chartreuse',
    'LeCaR': '#ed740b',
    'AdaptSize': '#669999',
    'LHD': '#336699',
    'GDWheel': '#000080',
    'Adaptive-TinyLFU': 'burlywood',
}
```

```python
trace_map = {
    'traceHK.tr': 'Akamai HK',
    'traceUS.tr': 'Akamai US',
    'wc200m.tr': 'Wikipedia',
    'wc400m.tr': 'Wikipedia',
    'memc_200m.tr': 'Memcachier',
    'msr_big_200m.tr': 'MSR Storage',
    'anonymous_200m.tr': 'Anonymous Application',
    'caida_cache_200m.tr': 'Caida',
}
```

```python
def cache_type_to_color(s: str):
    if s in color_map:
        return color_map[s]
    raise Exception(f'color {s} is not defined in color map')
```

```python
def top_algorithms(df, objective='byte_hit_rate'):
    mean_hit_rate = df.groupby(['cache_type', 'cache_size']).agg({objective: 'max'}).reset_index().groupby('cache_type').agg({objective: 'mean'})
    mean_hit_rate = mean_hit_rate.sort_values(by=objective, ascending=False)

    baseline_hit_rate = mean_hit_rate[~mean_hit_rate.index.isin(['GDBT'])]
    baseline_hit_rate = baseline_hit_rate.sort_values(by=objective, ascending=False)
    baseline_hit_rate = list(baseline_hit_rate.index.values[0:4])
    
    my_hit_rate = mean_hit_rate[mean_hit_rate.index.isin(['GDBT'])]
    my_hit_rate = my_hit_rate.sort_values(by=objective, ascending=False)
    my_hit_rate = list(my_hit_rate.index.values[0:1])
    
    return baseline_hit_rate+my_hit_rate
```

```python
def top_algorithm_all_csizes(df, objective='byte_hit_rate', k=4):
    top_algs = set()
    for l, grp in df1.groupby(['cache_type', 'cache_size']).agg({objective: 'max'}).reset_index().groupby('cache_size'):
        top_alg = set(grp[~grp.cache_type.isin(['GDBT', 'Belady'])].sort_values(by=[objective],  ascending=False).cache_type[0:k])
        top_algs = top_algs.union(top_alg)
    return top_algs
```

```python
# find top algorithms
df = common.load_reports()

filter_idx = (
    (df.trace_file == 'traceHK.tr') 
    & (df.uni_size == 0)
    & (~df.cache_type.isin(['Inf', 'BeladySample', 'LR'])) 
    & (25000000000 <= df.cache_size)
    & (df.cache_size <= 400000000000)
    & (df.n_warmup == 80000000)
#     & ((df.cache_type != 'LRUK') | (df.k == '4'))
)
df1 = df[filter_idx].copy()
top_algs = top_algorithm_all_csizes(df1, k=2)

filter_idx = (
    (df.trace_file == 'traceUS.tr') 
    & (df.uni_size == 0)
    & (~df.cache_type.isin(['Inf', 'BeladySample', 'LR'])) 
    & (25000000000 <= df.cache_size)
    & (df.cache_size <= 400000000000)
    & (df.n_warmup == 100000000)
    & ((df.cache_type != 'LRUK') | (df.k == '4'))
)
df1 = df[filter_idx].copy()
top_algs = top_algs.union(top_algorithm_all_csizes(df1, k=2))

filter_idx = (
    (df.trace_file == 'wc400m.tr') 
    & (df.uni_size == 0)
    & (~df.cache_type.isin(['Inf', 'BeladySample', 'LR'])) 
    & (25000000000 <= df.cache_size)
    & (df.cache_size <= 400000000000)
    & (df.n_warmup == 80000000)
    & ((df.cache_type != 'LRUK') | (df.k == '4'))
)

df1 = df[filter_idx].copy()
top_algs = top_algs.union(top_algorithm_all_csizes(df1, k=2))
```
